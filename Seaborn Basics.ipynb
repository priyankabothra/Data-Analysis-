## BoxPLot - Categorical Disribution plot

1. Find the median, 25th percentile and 75th percentile.
2. draw a box connecting all the lines for boxplot
3. Measure Inner Quartile Range(IQR) - distance between 25th percentile and 75th percentile range
4. Expand the whiskers on both sides with the range- 1.5 * IQR
5. the points left outside the whiskers are known and marked as outliers

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars.shape
```
