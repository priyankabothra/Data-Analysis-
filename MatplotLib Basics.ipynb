```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Simple plot with 4 numbers

```python
plt.plot([1,3,2,4])
plt.show()
```

## Points have x and y values; add title and axis labels 

```python
plt.plot([1,2,3,4], [1,4,9,16], 'r')
plt.title('Squares of Numbers', fontsize = 15, color = 'g') #title of the graph
plt.xlabel('Numbers : n', color = 'b') #x axis label
plt.ylabel('Squares : n^2', color = 'b') #y axis label
plt.show()
```

## Change figure size, plot red dots, set axis scales 

```python
plt.figure(figsize = (3,5)) #3 inches wide x 5 inches tall
plt.plot([1,2,3,4], [1,8,27,64],'rx') #red -circle(o)
plt.axis([0,8,0,80]) #[xmin, xmax, ymin, ymax]
plt.annotate('Cube of numbers', (3.5,45))
plt.show()
```

## Bar charts with 4 bars 

```python
plt.clf() #clear figure
x = np.arange(4)
y = [8.8, 5.2, 3.6, 5.9]
plt.xticks(x, ('Ankit', 'Hans', 'Joe', 'Flaco'))
#plt.bar(x, y)
# plt.bar(x, y, color = 'y')
plt.bar(x, y, color = ['b', 'r', 'lime', 'tan'])
plt.show()
```

## Two sets of 10 Random dots plotted

```python
d = {'Red O' : np.random.rand(10),
     'Grn X' : np.random.rand(10)}
df = pd.DataFrame(d)
df.plot(style = ['ro', 'gx'])
plt.show()
```

## Time series - six months of random floats

```python
ts = pd.Series(np.random.randn(180), index = pd.date_range('1/1/2021', periods = 180))
df = pd.DataFrame(np.random.randn(180,3), index = ts.index, columns=list('ABC'))
df.cumsum().plot()
#df.plot()
plt.show()
```

## Random dots in a scatter 

```python
N=50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = (30*np.random.rand(N)) ** 2 # 0 to 15 point radii
plt.scatter(x, y, s = sizes, c = colors, alpha = 0.5)
plt.show()
```

## Load csv file and show multiple chart types 

Or use plt.figlegend() to show a legend outside the plot area

```python
df = pd.read_csv('https://bit.ly/drinksbycountry')
df.head()
```

```python
plt.bar(df['continent'], df['beer_servings'], color = 'y')
plt.bar(df['continent'], df['spirit_servings'], color = 'lime')
plt.bar(df['continent'], df['wine_servings'], color = 'g')
plt.plot(df['continent'], df['total_litres_of_pure_alcohol'], color = 'b')
plt.legend()
plt.show()
```

## Subplots - 1

221 = top left subplot
222 = top right subplot
223 = bottom left subplot
224 = bottom right subplot

```python
fig = plt.figure() # create a new figure or plot
fig.suptitle('My Subplots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222, facecolor = 'g')
fig.add_subplot(223)
fig.add_subplot(224)
plt.show()
```

## Subplots - 2 

```python
fig,plots = plt.subplots(2, sharex = True) #share x axis
fig.suptitle('Sharing X axis')
x = range(0,200,5)
y = [n**0.8 for n in x]
plots[0].plot(x,y, color = 'r')
plots[1].scatter(x,y)
plt.show()
```
